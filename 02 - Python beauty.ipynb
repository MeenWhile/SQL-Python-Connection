```python
#Meen_Program

def Meen_Program():
    while True:
        my_cursor = my_db.cursor()

        # Create a string of MySQL commands
        sql_commands = Query_command()

        df = pd.read_sql(sql_commands, my_db)
        print("\n===== df =====")
        _ = display(df)

        if Export_Command == True:  
            # Execute one SQL command
            my_cursor.execute( sql_commands )

            # Get query results
            with open('ExportToCSV.csv',mode='w',encoding='utf-8',newline='') as f:
                f2 = csv.writer(f,delimiter=',')
                Export_Convert = []
                for i in Column_Want:
                    if i == 'c1' :
                        Export_Convert.append('product_id')
                    elif i == 'c2' :
                        Export_Convert.append('model')
                    elif i == 'c3' :
                        Export_Convert.append('name')
                    elif i == 'c4' :
                        Export_Convert.append('sub_id')
                    elif i == 'c5' :
                        Export_Convert.append('subcategory')
                    elif i == 'c6' :
                        Export_Convert.append('current_price')
                    elif i == 'c7' :
                        Export_Convert.append('raw_price')
                    elif i == 'c8' :
                        Export_Convert.append('currency')
                    elif i == 'c9' :
                        Export_Convert.append('discount')
                    elif i == 'c10' :
                        Export_Convert.append('likes_count')
                    elif i == 'c11' :
                        Export_Convert.append('is_new')
                    elif i == 'c12' :
                        Export_Convert.append('brand_id')
                    elif i == 'c13' :
                        Export_Convert.append('brand')
                    elif i == 'c14' :
                        Export_Convert.append('codCountry')
                    elif i == '*' :
                        Export_Convert.append('product_id')
                        Export_Convert.append('model')
                        Export_Convert.append('name')
                        Export_Convert.append('sub_id')
                        Export_Convert.append('subcategory')
                        Export_Convert.append('current_price')
                        Export_Convert.append('raw_price')
                        Export_Convert.append('currency')
                        Export_Convert.append('discount')
                        Export_Convert.append('likes_count')
                        Export_Convert.append('is_new')
                        Export_Convert.append('brand_id')
                        Export_Convert.append('brand')
                        Export_Convert.append('codCountry')
                        break
                f2.writerow(Export_Convert)
                for x in my_cursor.fetchall():
                    f2.writerow(x)
        What_Next = input('\tMake another table (Y/N)? : ').lower()
        if What_Next == 'y':
            continue
        elif What_Next == 'n':
            break
        else:
            print('\t\tsomething went wrong... automatically back to MENU.')
            break

            
            
def Query_command():
    
    #Column_Want
    print('\n===== Advanced Search =====')
    print('\nWhat Column do you want?')
    print('\t C1 = product_id \n\t C2 = model \n\t C3 = name \n\t C4 = sub_id \n\t C5 = subcategory \n\t C6 = current_price \n\t C7 = raw_price \n\t C8 = currency \n\t C9 = discount \n\tC10 = likes_count \n\tC11 = is_new \n\tC12 = brand_id \n\tC13 = brand \n\tC14 = codCountry ')
    global Column_Want
    while True:
        Column_Want = input('\tInput columns (seperated by comma , * = choose all columns): ').lower()
        Column_Want = Column_Want.split(',')
        Column_Command = Column_Convertion(Column_Want)
        if Column_Command != False:
            break

    
    
    #Row_Want
    print('\nWhat Row do you want?')
    while True:
        Row_Command = input('\tInput number (* = choose all rows): ')
        if Row_Command != '*':
            try:
                Row_Command = int(Row_Command)
                Row_Command = str(Row_Command)
                break
            except:
                print('\t\tsomething went wrong... change your input.')
                continue
        break

    
    
    #Condition_Want
    print('\nAdd conditions?')
    print('\t W = WHERE \n\t G = GROUP BY \n\t O = ORDER BY \n\t E = Export to CSV \n\t N = No add ')
    WHERE_Command = [' WHERE ']
    GROUP_Command = [' GROUP BY ']
    ORDER_Command = [' ORDER BY ']
    Condition_Want = input('\tInput condition: ').lower()
    while True:    
        global Export_Command

        #WHERE_Want
        if Condition_Want == 'w':
            WHERE_Command = [' WHERE ']
            WHERE_Want = input('\tInput WHERE condition (Ex. C6 < 25): ').lower()
            WHERE_Want = Column_Check(WHERE_Want)    
            for i in WHERE_Want:
                WHERE_Command.append(i)
            WHERE_Command = (''.join(WHERE_Command))
            print('\t\tConfirm WHERE condition.')
            print()
            Condition_Want = input('\tInput another condition or change previous condition: ').lower()
            continue

        
        #GROUP_Want
        elif Condition_Want == 'g':
            GROUP_Command = [' GROUP BY ']
            GROUP_Want = input('\tInput GROUP BY condition (Ex. C4): ').lower()
            GROUP_Want = Column_Check(GROUP_Want)
            for i in GROUP_Want:
                GROUP_Command.append(i)
            GROUP_Command = (''.join(GROUP_Command))
            print('\t\tConfirm GROUP BY condition.')
            print()
            Condition_Want = input('\tInput another condition or change previous condition: ').lower()
            continue
            

        #ORDER_Want
        elif Condition_Want == 'o':
            ORDER_Command = [' ORDER BY ']
            ORDER_Want = input('\tInput ORDER BY condition (Ex. C1, C10 DESC): ').lower()
            ORDER_Want = Column_Check(ORDER_Want)
            for i in ORDER_Want:
                ORDER_Command.append(i)
            ORDER_Command = (''.join(ORDER_Command))
            print('\t\tConfirm ORDER BY condition.')
            print()
            Condition_Want = input('\tInput another condition or change previous condition: ').lower()
            continue

            
       #Export_Want
        elif Condition_Want == 'e':
            Export_Command = True
            print('\t\tConfirm exporting to CSV.')
            print()
            Condition_Want = input('\tInput another condition or change previous condition: ').lower()
            continue

            
        #No_Add
        elif Condition_Want == 'n':
            break
            
            
        #Error Detect
        else:
            print('\t\tsomething went wrong... change your input.')
            Condition_Want = input('\tInput another condition or change previous condition: ').lower()
            continue
        
        
        
    #Generated_Command
    Generated_Command = ["SELECT"]
    Generated_Command.append(Column_Command)
    Generated_Command.append("FROM product JOIN brand ON product.brand_id=brand.No JOIN price ON product.product_id=price.product_id JOIN subcategory ON product.sub_id=subcategory.sub_id")
    if (WHERE_Command != [' WHERE ']):
        Generated_Command.append(WHERE_Command)
    if (GROUP_Command != [' GROUP BY ']):
        Generated_Command.append(GROUP_Command)
    if (ORDER_Command != [' ORDER BY ']):
        Generated_Command.append(ORDER_Command)
    if (Row_Command != '*'):
        Generated_Command.append("LIMIT")
        Generated_Command.append(Row_Command)
    Generated_Command.append(";")
    Generated_Command = ' '.join(Generated_Command)

    #print()
    #print(Generated_Command)
    
    return Generated_Command



def Column_Convertion(Want):
    Convert = []
    for i in Want:
        if i == 'c1' :
            Convert.append('product.product_id')
        elif i == 'c2' :
            Convert.append('product.model')
        elif i == 'c3' :
            Convert.append('product.name')
        elif i == 'c4' :
            Convert.append('product.sub_id')
        elif i == 'c5' :
            Convert.append('subcategory.subcategory')
        elif i == 'c6' :
            Convert.append('price.current_price')
        elif i == 'c7' :
            Convert.append('price.raw_price')
        elif i == 'c8' :
            Convert.append('price.currency')
        elif i == 'c9' :
            Convert.append('price.discount')
        elif i == 'c10' :
            Convert.append('product.likes_count')
        elif i == 'c11' :
            Convert.append('product.is_new')
        elif i == 'c12' :
            Convert.append('product.brand_id')
        elif i == 'c13' :
            Convert.append('brand.brand')
        elif i == 'c14' :
            Convert.append('product.codCountry')
        elif i == '*' :
            Convert.append('product.product_id')
            Convert.append('product.model')
            Convert.append('product.name')
            Convert.append('product.sub_id')
            Convert.append('subcategory.subcategory')
            Convert.append('price.current_price')
            Convert.append('price.raw_price')
            Convert.append('price.currency')
            Convert.append('price.discount')
            Convert.append('product.likes_count')
            Convert.append('product.is_new')
            Convert.append('product.brand_id')
            Convert.append('brand.brand')
            Convert.append('product.codCountry')
            break
        else:
            print('\t\tsomething went wrong... change your input.')
            Convert = False
            return Convert
    Convert = ','.join(Convert)
    return Convert



def Column_Check (Want):
    Check = list(Want)
    lst = []
    for i in range(len(Check)):
        if(Check[i] == 'c') and (i != len(Check)-1):
            if (Check[i+1] in ['1','2','3','4','5','6','7','8','9']):
                if (i != len(Check)-2) and (Check[i+2] in ['1','2','3','4','5','9','0']):
                    lst.append('c'+ Check[i+1] + Check[i+2])
                    Check[i] = ' ' + Column_Convertion(lst) + ' '
                    Check[i+1] = 'DUMMY'
                    Check[i+2] = 'DUMMY'
                    lst = []
                    continue
                lst.append('c'+ Check[i+1])
                Check[i] = ' ' + Column_Convertion(lst) + ' '
                Check[i+1] = 'DUMMY'
                lst = []
    Check = [x for x in Check if x != 'DUMMY']
    return Check
```

```python
#Nine_Program_Delete

def Nine_Program_Delete():
    my_cursor = my_db.cursor()
    
    print('\n=====Delete product=====')
    dl_pd_id = input('product_id to delete : ')
    sql_command_delete = f'SELECT product.product_id, \
                            product.model, \
                            product.name, \
                            product.sub_id, \
                            subcategory.subcategory, \
                            price.current_price, \
                            price.raw_price, \
                            price.currency, \
                            price.discount, \
                            product.likes_count, \
                            product.is_new, \
                            product.brand_id, \
                            brand.brand, \
                            product.codCountry \
                            FROM product  \
                            JOIN brand \
                            ON product.brand_id = brand.No \
                            JOIN price \
                            ON product.product_id = price.product_id \
                            JOIN subcategory \
                            ON product.sub_id = subcategory.sub_id \
                            WHERE product.product_id = {dl_pd_id};'
    df = pd.read_sql(sql_command_delete, my_db)
    print("\n===== df =====")
    _ = display(df)
    
    insert = input('Confirm your delete data (Y/N):')
    
    if insert == 'y' or insert == 'Y':
    
        sql_command_delete_product = f"""delete from product
                                            where product_id = {dl_pd_id}"""
        my_cursor.execute( sql_command_delete_product)
        my_db.commit()
        
        sql_command_delete_price = f"""delete from price
                                            where product_id = {dl_pd_id}"""
        my_cursor.execute( sql_command_delete_price)
        my_db.commit()
        print('Successfully deleted data from the database.')
    elif insert == 'n' or insert == 'N':
        print('Go back to MENU.')
    else:
        print('\t\tsomething went wrong... automatically back to MENU.')
```
